```python
def edit_distance(source, target):
    m = len(source)
    n = len(target)

    # Create DP table
    dp = [[0 for j in range(n + 1)] for i in range(m + 1)]

    # Initialize first column
    for i in range(m + 1):
        dp[i][0] = i

    # Initialize first row
    for j in range(n + 1):
        dp[0][j] = j

    # Fill the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):

            if source[i - 1] == target[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]

            else:
                insert = dp[i][j - 1]
                delete = dp[i - 1][j]
                replace = dp[i - 1][j - 1]

                dp[i][j] = 1 + min(insert, delete, replace)

    return dp
```

```python
# Function to print DP table
def print_table(dp, source, target):

    print("\nDP Table:")

    print("    #", *target)

    for i in range(len(dp)):

        if i == 0:
            print("#", *dp[i])
        else:
            print(source[i - 1], *dp[i])
```

```python
# Function to find sequence of operations
def operations(dp, source, target):

    i = len(source)
    j = len(target)

    steps = []

    while i > 0 or j > 0:

        # Match
        if i > 0 and j > 0 and source[i - 1] == target[j - 1]:

            steps.append("Match " + source[i - 1])
            i -= 1
            j -= 1

        # Replace
        elif i > 0 and j > 0 and dp[i][j] == dp[i-1][j-1] + 1:

            steps.append(
                "Replace " + source[i-1] + " with " + target[j-1]
            )

            i -= 1
            j -= 1

        # Delete
        elif i > 0 and dp[i][j] == dp[i-1][j] + 1:

            steps.append("Delete " + source[i-1])
            i -= 1

        # Insert
        else:

            steps.append("Insert " + target[j-1])
            j -= 1

    steps.reverse()

    print("\nSequence of Operations:")

    for step in steps:
        print(step)
```

```python
# Main Program

source = input("Enter source word: ")
target = input("Enter target word: ")

dp = edit_distance(source, target)

print_table(dp, source, target)

operations(dp, source, target)

print("\nMinimum Edit Distance:", dp[len(source)][len(target)])
```
